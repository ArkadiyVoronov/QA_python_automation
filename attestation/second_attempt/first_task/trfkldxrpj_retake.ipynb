# Задание

У руководства известного сервиса, занимающегося курьерской доставкой заказанных через интернет товаров появилась потребность: мониторить местоположение курьеров выполняющих доставку в данный момент. Для этого обратились к Вам с запросом на проектирование такой системы. Система должна представлять собой веб-сервис делающий две вещи:
- отображать местоположение всех выполняющих работу курьеров на карте
- обращаться http-запросами на выданные курьерам на работе мобильные устройства с целью  узнать их местоположение
    - при этом перед каждым запросом местоположения нужно сделать еще два технических запроса: узнать у устройства онлайн ли оно и узнать в базе выполняет ли курьер доставку прямо сейчас (чтобы не нагружать сеть и не отслеживать устройства курьеров вне рабочее время)

На первом этапе требуется только прототип системы. Модель данных в предполагаемой базе описывать НЕ нужно.

**_Задача - спроектировать систему классами. При выборе определённого шаблона проектирования объяснять выбор словами._**
